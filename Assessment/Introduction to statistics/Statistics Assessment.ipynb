1. What are the key properties of a Normal Distribution?

The key characteristics of distributions are:
 * Shape: It can either be bell-shaped or Symmetrical
 * Mean, Median and Mode: These are located at the center of the distribution
 * Spread: This is determined by the standard deviation
 * Total Area: This represents the total probability, i.e. 1. Hence, the total area under the curve is 1
